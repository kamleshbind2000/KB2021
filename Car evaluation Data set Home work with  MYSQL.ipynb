```python
"""1 . take a data set from attached URL https://archive.ics.uci.edu/ml/datasets/Car+Evaluation
2 . create a database called as cardataset
3 . create a table called as car with a column name given in dataset description
4 . Dump all the data into car table 
5 . try to check weather all the data is aviable inside your table or not 
6 . try to group all the data with COL1 AND count occurences of each and every record based on col1 value 
7 . Try to filter a record where col 3 value will be 4 .
8.  Try to update a col 3 value with 8 whereever you have value equal to 2 
9 . try to delete table 
10 .   Try to delete database"""
```

```python
# 1. create a Database as cardataset in  - MYSQL by Jupyter notebook


cursor.execute("CREATE database cardataset1")
```

```python
mydb.commit()
```

```python
cursor.execute("use cardataset1")
```

```python
# 2. Create table cardata with 7 attributes

cursor.execute("CREATE table cardataset1.cardata( buying VARCHAR(30), maint VARCHAR(30), doors VARCHAR(30), persons VARCHAR(30), lug_boot VARCHAR(30), safety VARCHAR(30), class VARCHAR(30))")
```

```python
mydb.commit()
```

```python
cursor.execute("show tables")
```

```python
cursor.fetchall()
```

```python
# Open car evaluation dataset in Jupyter notebook

file  = open ('car.data', 'r')
```

```python
# Display the car dataset 

import csv
with open ('car.data', 'r') as f :
    car_data = csv.reader(f, delimiter = '\n')
    for i in car_data:
        print(i)
```

```python
#4. Dump all data in the car table 


import csv

with open('car.data', 'r') as f :
            car_data=csv.reader(f,delimiter = '\n')
            for row in car_data:
                lst = row[0].split(',')
                cursor.execute("""INSERT INTO cardataset1.cardata values ("{0}","{1}","{2}","{3}","{4}","{5}","{6}")""".format(lst[0], lst[1], lst[2], lst[3], lst[4], lst[5], lst[6] ))
            mydb.commit()
```

```python
# 5. Check weather all the data is aviable inside your table or not 

cursor.execute("Select * from cardataset1.cardata")
```

```python
cursor.fetchall()
```

```python
# 6. Group all the data with COL1 AND count occurences of each and every record based on col1 value 

cursor.execute('Select buying, count(*)  from cardataset1.cardata GROUP BY buying')

```

```python
cursor.fetchall()
```

```python
# 7 . Filter a record where col3(door) value will be 4 


cursor.execute('Select *  from cardataset1.cardata WHERE doors=4 ')
```

```python
cursor.fetchall()
```

```python
#Try to update a col3(door) value with 8 where - ever you have value equal to 2 

try : 
    cursor.execute('UPDATE cardataset1.cardata SET doors = 8 WHERE doors = 2')

except Exception as e :
    print("The error is " , e)
```
